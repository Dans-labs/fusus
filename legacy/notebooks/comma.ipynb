```python
import numpy as np
from fusus.exe import Exe
```

```python
X = Exe(PREOCR_INPUT='test', OCR_INPUT='test/clean')
```

We show the band in which we look for the comma (the band is called `narrow`).
There are also `broad` and `inter` bands.
And there is a `main` band, directly derived from the histogram.

```python
rIm = X.start('qay_Page_1')
rIm.normalize()
rIm.histogram()
rIm.margins()
rIm.clean(mark="comma", line=None)
rIm.showBands("boxed", {"narrow"})
rIm.show(stage="boxed")
```

```python
rIm.showCleanInfo()
```
