# Explanation of source
The `fusus.csv` is a reduced form as created from the Lakhnawi PDF. It tries to strictly encapsulate the text of Fusus al-hikam. As the text is groomed and cleaned, `fusus.csv` is undergoing change, with the underlying code not preserved. This is especially true for columns right of 'word'. In particular 'short' is meant to be the cleanest. So far it has been stripped of diacritics, shaddas, and tatwīl.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import arabicABC as abc
```

```python
fusus = pd.read_csv('fusus.csv')
fusus = fusus.fillna('')
```

```python
fusus.head()
```

Per punctuatie bekijken wat er aan de hand is.

```python
punctMark = fusus[fusus['haspunct']==abc.MADDA_ABOVE]
```

```python
punctMark.head(60)
```

```python
fusus['haspunct'] = fusus['haspunct'].str.replace('ٓ','')
```

```python
def SetPunctuation(row):
    position = punctMark.iloc[row] 
    punct = position.haspunct
    short = position.short
    if short[-2:] == punct:
        fusus.iloc[position.name, fusus.columns.get_loc('punctAfter')] = punct
        fusus.iloc[position.name, fusus.columns.get_loc('short')] = short.replace(punct,'')
        fusus.iloc[position.name, fusus.columns.get_loc('haspunct')] = ""
```

```python
for i in range(0,len(punctMark)):
    SetPunctuation(i)
```

Welke punctuatie blijft over?

```python
puncs = fusus[fusus['haspunct']!=''].haspunct.unique().tolist()
```

```python
puncsLess = []
for punc in puncs:
    if '[' not in punc and ']' not in punc:
        puncsLess.append(punc)
print(puncsLess)
```

```python
for punc in puncsLess:
    times = fusus[fusus['haspunct']==punc].shape[0]
    print(punc + " " + str(times))
```
